```
#Human project (first training)
FOLD=['drive/My Drive/ResultadosHP',        #Original shape
#'drive/My Drive/ResultadosHP-ES',           #resize 
#Human project (second training)
'drive/My Drive/AnexosCNN/Resultados-01',   #Original shape
'drive/My Drive/AnexosCNN/Resultados-02',   #resize 
#Whitehall II imaging study
'drive/My Drive/AnexosCNN2/Resultados0',
'drive/My Drive/AnexosCNN2/Resultados1',
'drive/My Drive/AnexosCNN2/Resultados2',
'drive/My Drive/AnexosCNN2/Resultados4',
'drive/My Drive/AnexosCNN2/Resultados3',
'drive/My Drive/AnexosCNN2/Resultados5',
'drive/My Drive/AnexosCNN2/ResultadoX0',
'drive/My Drive/AnexosCNN2/ResultadoX1',
'drive/My Drive/AnexosCNN2/ResultadoX2',
'drive/My Drive/AnexosCNN2/ResultadoX3',
'drive/My Drive/AnexosCNN2/ResultadoX4',
'drive/My Drive/AnexosCNN2/ResultadoX5']
```

```
MODELS=['MODEL1-','LAYER1-','LAYER2-','LAYER3-','LAYER4-','LeNet-']
IMAGE=['AXI-','COM-','BLU-','PCA-','C_AXI-','C_COM-','C_BLU-','C_PCA-']
SLIC=['Axial-','Coronal-','Saggital-']
```

```
for PATH in FOLD:
  sample=[]
  models=[]
  images=[]
  slic=[]
  acc=[]
  vacc=[]
  loss=[]
  vloss=[]
  tacc=[]
  cont=0
  total=432
  #Models
  for i in range(6):
    #Images
    for j in range(8):
      #Slices
      for k in range(3):
        #Samples
        for l in range(3):
          namem=PATH+'/MAT/'+MODELS[i]+IMAGE[j]+SLIC[k]+str(l)+'.mat'
          namet=PATH+'/TEST/'+MODELS[i]+IMAGE[j]+SLIC[k]+str(l)+'.mat'
          if not path.exists(namem) or not path.exists(namem):
            break
          #Accuracy
          dat=sio.loadmat(namem)
          acc.append(np.max(dat['acc']))
          vacc.append(np.max(dat['val_acc']))
          loss.append(np.max(dat['loss']))
          vloss.append(np.max(dat['val_loss']))
          slic.append(SLIC[k][:-1])
          images.append(IMAGE[j][:-1])
          models.append(MODELS[i][:-1])
          sample.append(l)
          
          #Accuracy test
          dat=sio.loadmat(namet)
          tacc.append(dat['Accuracy test'][0][0])
          cont=cont+1
          print('\rProcess ', round(cont*100/total,2), '%...' , end ="")

  frame=pd.DataFrame({'sample': sample, 'models': models, 'images': images, 'slice': slic, 'acc': acc, 'val_acc': vacc, 'loss': loss, 'val_loss': vloss, 'test_acc': tacc})
  frame.to_csv('drive/My Drive/AnexoResul/'+PATH[-6:]+'.csv')
  print(' ')
  print(PATH)
  print(frame) 
```
